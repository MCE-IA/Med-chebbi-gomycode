```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
apriori(df, min_support=0.6)
```

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="lift", min_threshold=1.25)
```

```python
import pandas as pd
df=pd.read_csv('Market_basket_Optimisation.csv', header=None)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
column = []  
for i in range(0, 7501):  
    column.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

te = TransactionEncoder()
te_ary = te.fit(column).transform(column)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
apriori(df, min_support=0.0045)
```

```python
frequent_itemsets = apriori(df, min_support=0.0045, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.007)
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="lift", min_threshold=1.25)
```
