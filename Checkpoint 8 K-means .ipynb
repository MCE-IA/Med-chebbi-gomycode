```python
import pandas as pd
```

```python
cc_df=pd.read_csv("CC GENERAL.csv")
```

```python
cc_df.info()
```

```python
cc_df.head()
```

```python
cc_df.describe()
```

```python
cc_df.isna().sum()
```

```python
cc_df=cc_df.fillna(cc_df.median())
```

```python
cc_df.isna().sum()
```

```python
cc_df=cc_df.drop(labels="CUST_ID",axis=1)
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
clust_labels=model.fit_predict(cc_df)
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(111)
scatter=ax.scatter(cc_df['PAYMENTS'],cc_df['PURCHASES'],c=agglomerative[0],s=50)
ax.set_title('Agglomerative Clustering')
ax.set_xlabel('PAYMENTS')
ax.set_ylabel('PURCHASES')
plt.colorbar(scatter)
```

from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(cc_df, 'single')
plt.figure(figsize=(10,7))
plt.title("hierarchical clustering")
dendrogram(linked,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)

```python
from sklearn.cluster import KMeans
x = []
for i in range( 1, 30 ):
    kmeans = KMeans(n_clusters=i, init="k-means++", n_init=10, max_iter=300) 
    kmeans.fit_predict(cc_df)
    x.append( kmeans.inertia_ )
    
plt.plot( x)
plt.xlabel("Number of clusters")
plt.ylabel("x")
plt.show()
```

the best cluster is K-cluster=4

```python
kmeans = KMeans(n_clusters=4, init="k-means++", n_init=10, max_iter=300) 
cust_pred = kmeans.fit_predict(cc_df)
```

```python
plt.hist(cust_pred, bins=range(7))
plt.title('Customers per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Customers')
plt.show()
```
